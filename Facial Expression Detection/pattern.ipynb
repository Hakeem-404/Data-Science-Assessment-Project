## Pattern Matching

```python
import matplotlib.pylab as plt
import imageio.v2 as imageio
```

```python
original = imageio.imread('FacesSample/fig3166.jpg')
```

```python
plt.imshow(original)
plt.title("Search pattern in this image")
```

```python
template1 = original[10:25, 10:38]
template2 = original[10:25,10:22]
template3 = original[10:25,26:38]
```

```python
fig = plt.figure(figsize=(12,8))
f1 = plt.subplot(2,2,1)
f2 = plt.subplot(2,2,2)
f3 = plt.subplot(2,2,3)
f4 = plt.subplot(2,2,4)

f1.imshow(original)
f1.set_axis_off()
f1.set_title('Original')

f2.imshow(template1)
f2.set_axis_off()
f2.set_title('Both Eyes')

f3.imshow(template2)
f3.set_axis_off()
f3.set_title('Right Eye')

f4.imshow(template3)
f4.set_axis_off()
f4.set_title('Left Eye')
```

Perform pattern matching [Rectangle - Right Eye]

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from skimage.feature import match_template


correlation = match_template(original, template2)
xcoords = []
ycoords = []

for row in range(correlation.shape[0]):
    for col in range(correlation.shape[1]):
        if correlation[row, col] > 0.9:
            xcoords.append(col)
            ycoords.append(row)

# Plotting with rectangles
plt.imshow(original, cmap=plt.cm.gray)
plt.title("Found patterns")

for x, y in zip(xcoords, ycoords):
    rect = Rectangle((x, y), template2.shape[1], template2.shape[0], linewidth=2, edgecolor='m', facecolor='none')
    plt.gca().add_patch(rect)

plt.show()
```
