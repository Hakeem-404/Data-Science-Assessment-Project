#### B1 - Multiple linear regression

##### Test size = 0.2, Training data= 0.8

```python
df_ts = df.copy()
```

Spliting the data 

```python
y_pos = 10
model_features = [i for i in range(0,y_pos)]
X = df_ts.iloc[:, model_features]
Y = df_ts.iloc[:, y_pos]
from sklearn.model_selection import train_test_split 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state=0)
```

Create Model

```python
from sklearn.linear_model import LinearRegression
linear_regress = LinearRegression()
linear_regress.fit(X_train, Y_train)
intercept = linear_regress.intercept_
intercept
```

```python
coefficient = pd.DataFrame(linear_regress.coef_, X.columns, columns=['Coefficient'])
coefficient
```

Make predictions

```python
pred_y = linear_regress.predict(X_test)
df_result = pd.DataFrame({'Actual': Y_test, 'Predicted': pred_y})
df_result
```

Evaluate Model Accuracy

```python
from sklearn import metrics
print('R squared', metrics.r2_score(Y_test, pred_y))
print('MSE', f'{metrics.mean_squared_error(Y_test, pred_y):.3f}')
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, pred_y)))
```

Predicting Future Values

```python
pred_f = (lambda x: linear_regress.coef_* x + linear_regress.intercept_)
pred_f(10)
```

#### B2 - kNN Model

```python
df_kNN = df
df_kNN
```

##### Split data

```python
X = df_kNN.drop('Rating', axis=1) #X = df_kNN.iloc[:, :7].join(df.iloc[:, 9:])
Y = df_kNN['Rating'] #Y = df_kNN.iloc[:, 8]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.20, random_state=0)
```

##### Scaling and standardisation

```python
from sklearn.preprocessing import StandardScaler
```

```python
feature_scaler = StandardScaler()
feature_scaler.fit(X_train)
```

```python
X_train = feature_scaler.transform(X_train)
X_test = feature_scaler.transform(X_test)
```

##### Create model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, Y_train)
```

##### Make prediction

```python
pred_y = knn.predict(X_test)
pred_y
```

###### Confusion matrix

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(Y_test, pred_y))
```

###### Pecision and accuracy

```python
print(classification_report(Y_test, pred_y))
```

##### Hyperparameter tuning

```python
error = []
```

```python
for x in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors = x)
    knn.fit(X_train, Y_train)
    pred_x = knn.predict(X_test)
    error.append(np.mean(pred_x != Y_test))
```

```python
plt.figure(figsize = (12,6))
plt.plot(range(1,40), error, color = 'blue', linestyle = 'dashed', marker = 'o', markerfacecolor = 'blue', markersize = 10)
plt.title('Error Rate for k')
plt.xlabel('K values')
plt.ylabel('Mean Error')
```

Best k value believed to be 29

#### B3 - SVM model

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn.metrics import confusion_matrix, classification_report
```

```python
df_svm = df
df_svm
```

##### Split data

```python
X = df_svm.drop('Rating', axis=1) 
Y = df_svm['Rating']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.20, random_state=0)
```

##### Scaling and standardisation

```python
feature_scaler = StandardScaler()
feature_scaler.fit(X_train)
```

```python
X_train = feature_scaler.transform(X_train)
X_test = feature_scaler.transform(X_test)
```

##### LinearSVC Model

```python
svc_classifier = LinearSVC()
svc_classifier.fit(X_train, Y_train)
```

##### Make Predictions

```python
pred_y = svc_classifier.predict(X_test)
```

##### Evaluation model accuracy

###### Confusion matrix

```python
print(confusion_matrix(Y_test, pred_y))
```

###### Precision and Accuracy

```python
print(classification_report(Y_test, pred_y))
```

###### Hyperparameter tuning

```python
error = []
```

```python
for x in range(1, 40):
    param_grid = {'C': [x]}
    grid_search = GridSearchCV(svc_classifier, param_grid, cv=5, scoring='accuracy')
    grid_search.fit(X_train, Y_train)
    
    # Use the best model for prediction
    best_model = grid_search.best_estimator_
    pred_x = best_model.predict(X_test)
    
    # Calculate and append the error
    error.append(np.mean(pred_x != Y_test))
```

```python
x_values = range(1, 40)
error = error[:len(x_values)]  # Trim the error list if needed

plt.figure(figsize=(12, 6))
plt.plot(x_values, error, color='blue', linestyle='dashed', marker='o', markerfacecolor='blue', markersize=10)
plt.title('Error Rate for C')
plt.xlabel('C values')
plt.ylabel('Mean Error')
plt.show()
```

Best c value belived to be 32

#### B4 - k-means clustering

```python
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
```

```python
#Extracting the rating column because its the target
df_cluster = df.drop('Rating', axis=1)
```

```python
#standardinzing rating data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_cluster)
```

Create clusters

```python
kmeans = KMeans(n_clusters = 3, random_state = 11)
kmeans.fit(scaled_data)
```

Explore centroids and labels

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```

Visualise clusters and labels

```python
plt.scatter(scaled_data[:,0], scaled_data[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
```

##### Create estimator for dimensionality reduction using TSNE

```python
tsne = TSNE(n_components=2, random_state=11)
```

```python
sample = tsne.fit_transform(scaled_data)
```

```python
sample.shape
```

Visualise the reduced data

```python
graph = plt.scatter(sample[:,0], sample[:,1], cmap=('rainbow', 10))
colorbar = plt.colorbar(graph)
```

##### Create estimator for dimensionality reduction using PCA

```python
pca = PCA(n_components=2, random_state=11)
```

```python
sample = pca.fit_transform(scaled_data)
```

```python
sample.shape
```

Visualise the reduced data

```python
graph = plt.scatter(sample[:,0], sample[:,1], cmap=('rainbow', 10))
colorbar = plt.colorbar(graph)
```

As seen above, the clusters resulting from the TSNE methodology (line 143) are more distinct than those of the PCA methodology (line 147). In conclusion, the TSNE methodology is more appropriate in this instance.
